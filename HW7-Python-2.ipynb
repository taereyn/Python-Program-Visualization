```python
# import data
import pandas as pd
sheets_df = pd.read_excel("/Users/cerser/Downloads/EDAData.xlsx", sheet_name=None)
sheets_df
```

```python
#combine data to single excel sheet
allsheets = pd.concat(sheets_df.values())
allsheets
```

```python
# select Time for 0.00-1.00
allsheets1 = allsheets[allsheets['Time'] < 1.01]
allsheets1
```

```python
#norm function 
def norm_1(sp):
    sp_scaled = sp.copy()
    for column in sp_scaled.columns:
        sp_scaled[column] = (sp_scaled[column]-sp_scaled[column].abs().min())/((sp_scaled[column].abs().max()-sp_scaled[column].abs().min()))
    
    return sp_scaled
```

```python
# perform function to dataframe
newsheet = norm_1(allsheets1)
newsheet
```

```python
#drop columns to plot GSR Finger and Q Finger
newsheet1=newsheet.drop(['Q_Palm_Raw', 'Q_Wrist_Raw', 'Q_Foot_Raw', 'Q_Fingers',
       'Q_ftunder', 'Q_Palm','GSR_Palm_Raw', 'GSR_Wrist_Raw'], axis=1)
newsheet1
```

```python
# Plot first 25 subjects overlapping 
import matplotlib.pyplot as plt
ax = newsheet1.plot()
fig= ax.get_figure()
fig.savefig('/Users/cerser/Downloads/hw7.pdf')
```
